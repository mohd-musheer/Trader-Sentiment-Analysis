```python
from pathlib import Path

import pandas as pd


base_dir = Path.cwd()
fear_greed_path = base_dir / "fear_greed_index.csv"
historical_data_path = base_dir / "historical_data.csv"

missing_files = [path.name for path in [fear_greed_path, historical_data_path] if not path.exists()]
if missing_files:
    raise FileNotFoundError(f"Missing required file(s): {', '.join(missing_files)}")

fear_greed = pd.read_csv(fear_greed_path)
historical_data = pd.read_csv(historical_data_path)

print("fear_greed_index shape:", fear_greed.shape)
print("historical_data shape:", historical_data.shape)
print("\nFear & Greed preview:")
print(fear_greed.head())
print("\nHistorical data preview:")
print(historical_data.head())
```
